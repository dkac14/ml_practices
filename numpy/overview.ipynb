## Loading libraries

```python
import numpy as np
```

## list -> numpy.ndarray

```python
# Python List
a = [1, 2, 3, 4, 5]
b = [6, 7, 8, 9, 10]

# Converting in ndarray
a_np = np.array(a)
b_np = np.array(b)

print("a: ", a)
print(type(a))

print(f"a_np: {a_np}")
print(type(a_np))
```

## numpy.ndarray -> list

```python
# .tolist to convert ndarray -> python list
b_list = b_np.tolist()

print(b_list)
print(type(b_list))
```
