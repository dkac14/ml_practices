## Importing Libraries

```python
import numpy as np
import pandas as pd
```

## Loading dataset

```python
URL = "C:\\Users\\Gateway\\Documents\\GitHub\\ml_practices\\numpy\\GHCND_sample_csv.csv"
raw = pd.read_csv(URL)

raw.head(7) #View the first five rows
```

## Visualizing Data from dataset

### Week 1 - TMax, Tmin, Prcp

```python
week_one_tmax = raw.iloc[:7]["TMAX"]
week_one_tmin = raw.iloc[:7]["TMIN"]
week_one_prcp = raw.iloc[:7]["PRCP"]

print(f"Week 1 TMAX:\n{week_one_tmax}")
print(f"Week 1 TMIN:\n{week_one_tmin}")
print(f"Week 1 PRCP:\n{week_one_prcp}")
```

## Length of One-Dimensional Array -> len()

```python
# With Function: len()
print("Size of week 1 TMAX: ", len(week_one_tmax))

# With attribute: shape
print("Size of week 1 TMAX: ", week_one_tmax.shape)
```

## Converting TMAX & TMIN in Celsius
Since TMAX and TMIN are measured in tenths of degrees Celsius, we can divide by 10 to convert to degrees Celsius. PRCP should be divided by 10 to convert to millimeters.

```python
week_one_tmax = week_one_tmax / 10
week_one_tmin = week_one_tmin / 10
```

### Daily Temperature Range

```python
week_one_trange = week_one_tmax - week_one_tmin
print("Week 1  TRANGE (in Celsius):\n", week_one_trange)
```

## Build-in Functions

```python
print("Hottest day in Week 1:", np.max(week_one_tmax))
print("Total rainfall in Week 1:", np.sum(week_one_prcp))
print("Average TMAX in Week 1:", np.mean(week_one_tmax)) # mean is the average
print("Standard deviation of TMAX in Week 1:", np.std(week_one_tmax)) # std is the standard deviation
```
